#### 1. Collaborative Filtering using Matrix Factorization (SVD)

```python
# Load training data
reader = Reader(rating_scale=(1, 10))
data = Dataset.load_from_df(train_df_clean[['user_id', 'anime_id', 'rating']], reader)
```

```python
# Split into training and validation sets
trainset, testset = train_test_split(data, test_size=0.2)
```

```python
# Train SVD model
model_svd = SVD()
model_svd.fit(trainset)
```

```python
# Predict and evaluate
predictions = model_svd.test(testset)
rmse_svd = accuracy.rmse(predictions, verbose=True)
```

```python
# Collect predictions into a list
results = []
for user_id, anime_id, actual_rating, estimated_rating, _ in predictions:
    results.append({'user_id': user_id, 'anime_id': anime_id, 'actual_rating': actual_rating, 'predicted_rating': estimated_rating})

# Convert list of results into a DataFrame
predictions_df = pd.DataFrame(results)

# View the predictions for each user-anime pair
print(predictions_df.head())  # Display the first few rows
```

```python
def recommend_collaborative(user_id, svd_model, anime_df, train_df, top_n=10):
    # Get the user's previously rated anime
    user_history = train_df[train_df['user_id'] == user_id]['anime_id'].values
    
    # Predict ratings for all anime IDs
    recommendations = []
    for anime_id in anime_df['anime_id'].values:
        if anime_id not in user_history:  # Exclude watched titles
            pred_rating = svd_model.predict(user_id, anime_id).est
            recommendations.append({'anime_id': anime_id, 'predicted_rating': pred_rating})
    
    # Sort by predicted rating
    recommendations_df = pd.DataFrame(recommendations).sort_values(by='predicted_rating', ascending=False)
    
    return recommendations_df.head(top_n)

# Example usage
recommended_anime = recommend_collaborative(user_id=1, svd_model=model_svd, anime_df=anime, train_df=train)
print(recommended_anime)
```

```python
# Content-Based Filtering (Srarahs code)

tfidf = TfidfVectorizer(stop_words = 'english')
tfidf_matrix = tfidf.fit_transform(anime_df['genre'])
content_similarity = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Content-based recommendations
def get_content_recommendations(anime_id, top_n=10):
    idx = anime[anime_df['anime_id'] == anime_id].index[0]
    sim_scores = list(enumerate(content_similarity[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:top_n+1]
    anime_indices = [i[0] for i in sim_scores]
    return anime[['anime_id', 'name']].iloc[anime_indices]
```

#### 3. Hybrid Approach: Weighted Combination of SVD and Content-Based Similarity

```python
#Sarahs code
def hybrid_recommendations(user_id, anime_id, top_n=10, alpha=0.5):
    idx = anime_df_clean[anime_df_clean['anime_id'] == anime_id].index[0]
    sim_scores = list(enumerate(content_similarity[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:top_n*2]  # Take more in case some aren't rated

    recommendations = []
    for i, content_score in sim_scores:
        other_anime_id = anime_df_clean.iloc[i]['anime_id']
        collab_score = predict_rating(user_id, other_anime_id)
        final_score = alpha * collab_score + (1 - alpha) * content_score
        recommendations.append((other_anime_id, anime.iloc[i]['name'], final_score))

    recommendations = sorted(recommendations, key=lambda x: x[2], reverse=True)
    return recommendations[:top_n]
```

```python
#Ammaarah's code
# Get content-based similarity score
def content_score(user_id, anime_id, cosine_sim, anime_df, user_history):
    if anime_id not in anime_df['anime_id'].values:
        return 0  # If anime is not in dataset
    similar_anime = get_recommendations(anime_id, cosine_sim, anime_df)
    similar_ids = similar_anime['anime_id'].values
    user_ratings = user_history[user_history['anime_id'].isin(similar_ids)]
    return user_ratings['rating'].mean() if not user_ratings.empty else 0

# Combine SVD and content-based scores
def hybrid_score(user_id, anime_id, svd_model, cosine_sim, anime_df, train_df):
    pred_svd = svd_model.predict(user_id, anime_id).est
    user_history = train_df_clean[train_df_clean['user_id'] == user_id]
    pred_content = content_score(user_id, anime_id, cosine_sim, anime_df, user_history)
    return 0.7 * pred_svd + 0.3 * pred_content  # Weighted combination
```
